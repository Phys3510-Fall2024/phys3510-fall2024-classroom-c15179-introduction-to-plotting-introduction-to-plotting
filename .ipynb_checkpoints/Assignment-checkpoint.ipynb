```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider
```

```python
def potential_energy(Q1, Q2, K1=1.0, K2=1.0, C=0):
    return 0.5*K1*Q1**2 + 0.5*K2*Q2**2 +C* Q1*Q2**2 #following the README.md file

```

```python
#define meshes
q1= np.linspace(-1,1,100)
q2=q1
Q1, Q2= np.meshgrid(q1,q2)

#potential_energy(Q1,Q2)
```

```python
#plot contours with matplotlib contourf
plt.figure(figsize=(4,3))
plt.contourf(Q1,Q2, potential_energy(Q1,Q2, K1=0.2, K2=1.0,C=-0.7), levels=30)
plt.colorbar(label='potential energy')
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.xlabel('$Q_1$')
plt.ylabel('$Q_2$')


plt.show()
```

```python
#define plotting function
def plot_potential(K1=1.0, K2=1.0, C=0, levels=10,n_points=50):
    #define mesh and meshgrid to plot
    q1= np.linspace(-1,1,n_points)
    q2=q1
    Q1, Q2= np.meshgrid(q1,q2)
    Z= potential_energy(Q1,Q2, K1, K2,C)

    #construct contour plot
    plt.figure(figsize=(4,3))
    plt.contourf(Q1,Q2, Z, levels)
    plt.colorbar(label='potential energy')
    plt.xlim([-1,1])
    plt.ylim([-1,1])
    plt.xlabel('$Q_1$')
    plt.ylabel('$Q_2$')
    plt.show()
```

```python
plot_potential(K1=1.0, K2=1.0, C=0, levels=10,n_points=50)
```

```python
interact(plot_potential,
    K1=FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K1'),
    K2=FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K2'),
    C=FloatSlider(value=1.0, min=-5.0, max=5.0, step=0.1, description='C'),
    levels=IntSlider(value=15, min=1, max=50, step=5, description='#contours'),
    n_points=IntSlider(value=50, min=1, max=50, step=5, description='#points'))
```

#### 1. As you increase $C$, how do the low-energy contours start to change?

Increasing $C$ made the low energy contours increase in potential energy after the $C$ changes substantially. A small change in $C$ makes more of a guitar pick shape of the contour. As $C$ increases, the energy landscape is more pronouced, leading to contours that expand rapidly in the $Q_1$ direction for given values of $Q_2$. There are more complex shapes, and the interaction between $Q_1$ and $Q_2$ leads to more pronounced contours that reflect attraction or repulsion. 

#### 2. Does the force on $Q_1$ depend on the direction of $Q_2$?

The term $CQ_1Q_2^2$ is indicative that the force related to $Q_1$ is directly influenced by what $Q_2$ is. If $Q_1$ has a directional component, as a vector, then $Q_2's$ direction would depend on it. If just evaluating the magnitudes, using scalars, then the effect is purely based on its magnitude. 

#### 3. What do you expect the influence of driving $Q_2$ to be on the motion of $Q_1$?

If $Q_2$ increases, the force on $Q_1$ increases due to the $Q_2^2$ term. This can results in greater acceleration, as $force = ma$. If $Q_2$ has a directional component, its influend could lead to changes in the trajectory/ path of $Q_1$. If both $Q_1$ and $Q_2$ are in motion together, their relative accelerations could affect the other, based on their magnitude. This could affect their oscillations. 

#### 4. If $Q_1$ is kicked away from equilibrium, what do you expect its influence to be on $Q_2$?

As $Q_1$ is away from equilibrium, it could create forces that pull or push $Q_2$ in response. The resulting forces may lead to a new equilibrium point for the system or cause $Q_2$ to stabilize in a different position.
