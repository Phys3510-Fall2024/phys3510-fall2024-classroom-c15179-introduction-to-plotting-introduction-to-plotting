## Plotting with `matplotlib`

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the potential energy
def potential_energy(x, k=1.0):
    return 0.5 * k * x**2 #k is the spring constant
    
potential_energy(3,10)
```

```python
# define mesh for plotting
x= np.linspace(-5,5,3) # (x_min, x_max, npts)

#define y values from x list
y= potential_energy(x)

print(np.transpose([x,y])) #testing the output
```

```python
#plot the function

plt.plot(x,y) #only used three points, so it does not look like a parabola
```

```python
# define mesh for plotting
x= np.linspace(-5,5,100) # (x_min, x_max, npts)

#define y values from x list
y= potential_energy(x)

#print(np.transpose([x,y])) #testing the output
```

```python
plt.figure(figsize=(5,4)) #figure size in inches

plt.plot(x,y, color="black", linewidth=0.5) 

#add axes labels
plt.xlabel("displacement (units)")
plt.ylabel("potential energy (units)")

#change limits of x and y
plt.xlim([-4,4])
plt.ylim([-1,10])

plt.show() #combine all elements of the plot
```

## Contour plots

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the potential energy
def potential_energy_2D(x,y, kx=1.0, ky=1.0):
    return 0.5 * (kx * x**2 + ky * y**2) #kx and ky are spring constants

potential_energy_2D(2,1)
```

```python
# define mesh for plotting
x= np.linspace(-5,5,100) # (x_min, x_max, npts)
y= np.linspace(-5,5,100) # (y_min, y_max, npts)

#combine 1D meshes into meshgrid
X,Y = np.meshgrid(x,y)

#define Z values
Z=potential_energy_2D(X,Y)

```

```python
# construct the contour plot
plt.figure(figsize=(4,3)) #figure size in inches

plt.contourf(X,Y,Z, levels=10)

plt.xlim([-4,4])
plt.ylim([-4,4])

plt.xlabel("x-displacement (units)")
plt.ylabel("y-displacement (units)")

#add legend
plt.colorbar(label="potential energy")

plt.show()
```
