```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider
```

```python
def potential_energy(Q1, Q2, K1=1.0, K2=1.0, C=0):
    return 0.5*K1*Q1**2 + 0.5*K2*Q2**2 +C* Q1*Q2**2 #following the README.md file

```

```python
#define meshes
q1= np.linspace(-1,1,100)
q2=q1
Q1, Q2= np.meshgrid(q1,q2)

#potential_energy(Q1,Q2)
```

```python
#plot contours with matplotlib contourf
plt.figure(figsize=(4,3))
plt.contourf(Q1,Q2, potential_energy(Q1,Q2, K1=0.2, K2=1.0,C=-0.7), levels=30)
plt.colorbar(label='potential energy')
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.xlabel('$Q_1$')
plt.ylabel('$Q_2$')


plt.show()
```

```python
#define plotting function
def plot_potential(K1=1.0, K2=1.0, C=0, levels=10,n_points=50):
    #define mesh and meshgrid to plot
    q1= np.linspace(-1,1,n_points)
    q2=q1
    Q1, Q2= np.meshgrid(q1,q2)
    Z= potential_energy(Q1,Q2, K1, K2,C)

    #construct contour plot
    plt.figure(figsize=(4,3))
    plt.contourf(Q1,Q2, Z, levels)
    plt.colorbar(label='potential energy')
    plt.xlim([-1,1])
    plt.ylim([-1,1])
    plt.xlabel('$Q_1$')
    plt.ylabel('$Q_2$')
    plt.show()
```

```python
plot_potential(K1=1.0, K2=1.0, C=0, levels=10,n_points=50)
```

```python
interact(plot_potential,
    K1=FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K1'),
    K2=FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K2'),
    C=FloatSlider(value=1.0, min=-5.0, max=5.0, step=0.1, description='C'),
    levels=IntSlider(value=15, min=1, max=50, step=5, description='#contours'),
    n_points=IntSlider(value=50, min=1, max=50, step=5, description='#points'))
```
